## Class

### Define a Class

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    def hello(self):
        print("Hello " + self.name + "!")
    def goodbye(self):
        print("Good-bye " + self.name + "!")
```

### Call the Class

```python
m = Man("David")
m.hello()
m.goodbye()
```
