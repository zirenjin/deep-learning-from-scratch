## Step Function

```python
import numpy as np

def step_function(x):
    y = x > 0
    return y.astype(int)

x = np.array([-1.0,1.0,2.0])
print(x)
y = step_function(x)
print(y)
```

## Sigmoid Function

```python
import numpy as np

def sigmoid(x):
    return 1 / (1+np.exp(-x))

x = np.array([-1.0,1.0,2.0])
y = sigmoid(x)
print(y)
```

## ReLU Function

```python
import numpy as np

def relu(x):
    return np.maximum(0,x)

x = np.array([-1.0,1.0,2.0])
print(relu(x))
```

## Graph of Step Function

```python
import matplotlib.pylab as plt
import numpy as np

def step_function(x):
    return np.array(x>0,dtype=int)

x = np.arange(-5.0,5.0,0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```
